```python
'''Task 2) Identify the columns that have null values and perform missing value treatment'''
data.isnull().sum()
data['tagline'].fillna('No tagline', inplace=True)
data['homepage'].fillna('No homepage', inplace =True)
data.dropna(inplace=True)
```

```python
'''Task 3)Display the movie categories that have a budget greater than $220000.'''
x = data[data['budget']>220000]
y = x[['genres']]
pd.DataFrame(y)
```

```python
'''Task 4)Display the movie categories where the revenue is greater than $961000000.'''
x = data[data['revenue']>961000000]
y = x[['genres']]
pd.DataFrame(y)
```

```python
'''Task 5)In the dataset,there are some movies for which the budget and revenue colum have the value 0,which mean unknown value.
Remove the rows with the value 0 from both the budget and revenue columns.'''
filtered_data = data[(data['budget']!=0) & (data['revenue']!=0)]
filtered_data
```

```python
'''Task 6) List the top 10 movies with the highest revenues and the top 10 movies with the least budget.'''
filtered_data = data[(data['revenue']!=0) & (data['budget']!=0)]
top_highest_revenue=filtered_data.sort_values(by='revenue',ascending=False).head(10)
top_highest_revenue_movies = top_highest_revenue[['title','revenue']].reset_index(drop=True)
top_least_budgets=filtered_data.sort_values(by='budget',ascending=True).head(10)
top_least_budget_movies =top_least_budgets[['title','budget']].reset_index(drop=True)
print("Top 10 highest revenue Movies")
print(top_highest_revenue_movies)
print("Top 10 least Budget Movies")
print(top_least_budget_movies)
```

```python
'''Task7) How are popularities of movies related with the movie budgets? 
Are they correlated or totally uncorrelated with each other? 
Write the interpretation of your analysis'''
filtered_data = data[data['budget']!=0]
plt.scatter(filtered_data['popularity'],filtered_data['budget'],alpha=0.5)
plt.title('Correlation between Budget vs Popularity')
plt.xlabel('Total_Budget')
plt.ylabel('Popularity')
plt.show()
```

```python
'''Task 8) Identify and display the names of all production companies along with the number of times they appear in dataset.'''
x= data['production_companies'].value_counts()
pd.DataFrame(x)
```

```python
'''Task 9)Display the names of the top 25 production companies based on the number of movies they have produced in descending 
order of the number of movies produced.'''
filtered_data = data[data['production_companies']!='[]']
x = filtered_data['production_companies'].value_counts().sort_values(ascending = False).head(25)
pd.DataFrame(x)
```

```python
'''Task 10)Sort the data in descending order based on revenue and filter the top 500 movies.
Find the measures of central tendency for the following columns using the filtered data.
a) budget	b) revenue	c) runtime'''
sort_data = data.sort_values(by='revenue',ascending=False).head(500)
revenue_mean = sort_data['revenue'].mean()
revenue_mode = sort_data['revenue'].mode()[0]
revenue_median = sort_data['revenue'].median()

budget_mean = sort_data['budget'].mean()
budget_mode = sort_data['budget'].mode()[0]
budget_median = sort_data['budget'].median()

runtime_mean = sort_data['runtime'].mean()
runtime_mode = sort_data['runtime'].mode()[0]
runtime_median = sort_data['runtime'].median()

print('Revenue Mean = ',revenue_mean)
print('Revenue Mode = ',revenue_mode)
print('Revenue Median = ',revenue_median)

print('Budget Mean = ',budget_mean)
print('Budget Mode = ',budget_mode)
print('Budget Median = ',budget_median)

print('Runtime Mean = ',runtime_mean)
print('Runtime Mode = ',runtime_mode)
print('Runtime Median = ',runtime_median)
```

```python
'''Task 11)Identify and display the names of the movies along with their run times for those movies that have above average 
runtime, using the data from the previous task.'''
average_runtime = sort_data['runtime'].mean()
average_runtime_movies = sort_data[sort_data['runtime']>average_runtime]
average_runtime_movies[['title','runtime']]
```
